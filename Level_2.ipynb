## Analysis of English Premier League Player Statistics

This notebook explores a dataset of English Premier League players, focusing on calculating basic summary statistics such as mean, median, mode, minimum, and maximum for key variables. We also interpret these statistics to understand the players' central tendency, variability, and distribution and make initial observations on player demographics and performance.


```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/AFRONEX Internships/Players Stat.csv')

# Display the first few rows of the dataset
data.head()
```

### Summary Statistics

We will calculate and display the summary statistics for key numerical variables: `Age`, `Appearances`, and `Wins`.


```python
# Calculating summary statistics for Age
age_describe = data['Age'].describe()

# Calculating summary statistics for Appearances
appearances_describe = data['Appearances'].describe()

# Calculating summary statistics for Wins
wins_describe = data['Wins'].describe()

# Display the summary statistics
print("Age Statistics:\n", age_describe)
print("\nAppearances Statistics:\n", appearances_describe)
print("\nWins Statistics:\n", wins_describe)
```

```python
# Calculate summary statistics for Age
age_mean = data['Age'].mean()
age_median = data['Age'].median()
age_mode = data['Age'].mode()[0]  # mode() returns a Series, get the first mode
age_min = data['Age'].min()
age_max = data['Age'].max()

# Display the summary statistics
print(f"Mean Age: {age_mean}")
print(f"Median Age: {age_median}")
print(f"Mode Age: {age_mode}")
print(f"Minimum Age: {age_min}")
print(f"Maximum Age: {age_max}")
```

```python
# Calculate summary statistics for Appearances
appearances_mean = data['Appearances'].mean()
appearances_median = data['Appearances'].median()
appearances_mode = data['Appearances'].mode()[0]  # mode() returns a Series, get the first mode
appearances_min = data['Appearances'].min()
appearances_max = data['Appearances'].max()

# Display the summary statistics
print(f"Mean Appearances: {appearances_mean}")
print(f"Median Appearances: {appearances_median}")
print(f"Mode Appearances: {appearances_mode}")
print(f"Minimum Appearances: {appearances_min}")
print(f"Maximum Appearances: {appearances_max}")
```

```python
# Calculate summary statistics for Wins
wins_mean = data['Wins'].mean()
wins_median = data['Wins'].median()
wins_mode = data['Wins'].mode()[0]  # mode() returns a Series, get the first mode
wins_min = data['Wins'].min()
wins_max = data['Wins'].max()

# Display the summary statistics
print(f"Mean Wins: {wins_mean}")
print(f"Median Wins: {wins_median}")
print(f"Mode Wins: {wins_mode}")
print(f"Minimum Wins: {wins_min}")
print(f"Maximum Wins: {wins_max}")
```

### Data Distribution Visualizations

Visualizing the distribution of `Age`, `Appearances`, and `Wins` to better understand the central tendency and variability.


```python
# Histogram for Age
plt.figure(figsize=(10, 6))
plt.hist(data['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for Appearances
plt.figure(figsize=(10, 6))
plt.hist(data['Appearances'].dropna(), bins=20, color='green', edgecolor='black')
plt.title('Distribution of Player Appearances')
plt.xlabel('Appearances')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for Wins
plt.figure(figsize=(10, 6))
plt.hist(data['Wins'].dropna(), bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Player Wins')
plt.xlabel('Wins')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Observations and Insights

Based on the calculated statistics and visualizations, here are some key insights and observations regarding the players in the English Premier League:


#### Age Observations:
- The distribution of ages can help identify the career lifecycle typical in the league.
- A close mean and median suggest a symmetric age distribution among players.

#### Appearances Observations:
- A right-skewed distribution of appearances may indicate that while most players appear in fewer games, a few have very high participation rates.

#### Wins Observations:
- Players with high wins might be key contributors to their teams, suggesting their significant impact on match outcomes.
